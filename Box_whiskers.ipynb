```python
#creating a bar plot using sns


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\Administrator\Desktop\Python Test Data\NPTEL\Toyota.csv', index_col= 0)
data.dropna(axis= 0, inplace= True)
```

```python
sns.boxplot(y= data['Price'])
```

```python
#for categorical vs numerical

sns.boxplot(x= 'FuelType', y= 'Price', data= data)
```

```python
#boxplot for Price vs Fueltype and Automatic


sns.boxplot(x= 'FuelType', y= 'Price',hue= 'Automatic',  data= data)
```

```python
#splitting plottng window

f,(ax_box, ax_hist)= plt.subplots(2, gridspec_kw= {'height_ratios': (0.5, .5)})

#adding two plots

sns.boxplot(x= data['Price'], ax= ax_box)
sns.distplot(data['Price'],ax= ax_hist, kde= False, bins=5)
```
