```python
# Data Visualisation using matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading file
data = pd.read_csv(r'C:\Users\Administrator\Desktop\Python Test Data\NPTEL\Toyota.csv',index_col=0)

#viewing basic data
data.head(5)
data.tail(5)
```

```python
#removing missing values from dataframe
data.dropna(axis=0, inplace= True)
```

```python
data.head(25)
```

```python
#creating a scatter plot
plt.scatter(data['Age'], data['Price'], c ='red')
```

```python
# from the following plot we see that there is a negative correlation between age and price of cars.
```

```python
#making the plot more descriptive
plt.scatter(data['Age'], data['Price'], c= 'blue')
plt.title('Scatter plot of Price vs Age of cars')
plt.xlabel('Age(months)')
plt.ylabel('Price(euros)')
plt.show()
```

```python
#scatter plot of Kms vs Age

plt.scatter(data['Age'], data['KM'], c= 'red')
plt.title('Scatter plot of kms travelled vs age of car')
plt.xlabel('Age')
plt.ylabel('KMs travelled')
plt.show()
```

```python
# there is a positive correlation between age and kms travelled by a car
```

```python
# Scatter plot between Price vs kms travelled

plt.scatter(data['KM'], data['Price'], c= 'black')
plt.title('Scatter plot of Price vs Kms travelled')
plt.xlabel('KMs travelled')
plt.ylabel('Price')
plt.show()
```
